```python
%matplotlib inline

# Import dependencies.
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
#Object Oriented Approach
# Create the plot and add a label for the legend.
fig, ax = plt.subplots()

# Create labels for the x and y axes.
ax.set_xlabel("Date")
ax.set_ylabel("Fare")
ax.plot(x_axis, y_axis, color = "green", marker = "D")

# # Set the y limit between 0 and 45.
ax.set_ylim(0, 45)
# # Create a title.
ax.set_title("PyBer Fare By Month")
# # Add a grid.
ax.xaxis.grid()
ax.yaxis.grid()
# # Add the legend.
fig.legend("Boston")
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.barh(x_axis, y_axis, color="magenta", label='Boston')
plt.gca().invert_yaxis()
plt.title("PyBer Fare by Month")
plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
ax.bar(x_axis, y_axis, color="black", label='New York')
plt.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.invert_yaxis()
ax.barh(x_axis, y_axis, color="cyan", label='Chicago')
plt.legend()
plt.show()
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(y_axis, x_axis, color = "red", label='Chicago')

# Create labels for the x and y axes.
plt.gca().invert_yaxis()
plt.ylabel("Date")
plt.xlabel("Fare($)")
# Set the y limit between 0 and 45.
plt.xlim(0, 50)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
# plt.grid()
# Add the legend.
plt.legend()
```

```python
#Increase the size of the dots by a factor of 3
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
#Rerun scatter plot with the larger icons
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
#Include the condition in the plot instructions
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
#Using Object Oriented Plotting
fig, ax = plt.subplots()
ax.invert_yaxis()
ax.scatter(y_axis, x_axis, c = "skyblue", s = [i * 5 for i in y_axis] , edgecolors= "black", label = 'Boston')
ax.set_xlim(0, 50)
ax.set_ylabel("Date")
ax.set_xlabel("Fare ($)")
ax.set_title("PyBer Fares by Month")
plt.legend()
plt.show()
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python

fig = plt.figure()
fig.set_figheight(8)
fig.set_figwidth(8)
fig, ax = plt.subplots()
ax.pie(y_axis,explode = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0), autopct='%1.1f%%',shadow=True, labels=x_axis, rotatelabels = True, startangle=80)
plt.show()
```
